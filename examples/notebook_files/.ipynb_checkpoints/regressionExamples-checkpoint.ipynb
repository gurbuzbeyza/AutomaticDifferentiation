```python
import autograd.numpy as np
import matplotlib.pyplot as plt
import time
```

```python
start_time = time.time()
n_samples = 100
X_data = np.linspace(1, 200, n_samples).reshape((n_samples, 1))
Y = 4*np.cos(5*X_data) - 7*np.sin(2*X_data) + 6 + 5*np.random.randn(n_samples).reshape((n_samples, 1))
# Y = Y.reshape((n_samples, 1))
X = np.hstack([np.cos(5*X_data), np.sin(2*X_data), np.ones((n_samples, 1))])
f = open('../txt_files/regexds.txt', 'w+')
f2 = open('../txt_files/regexlb.txt', 'w+')
for x in X.tolist():
    f.write(str(x[0]) + ' ' + str(x[1]) + ' ' + str(x[2]) + '\n')
for y in Y.tolist():
    f2.write(str(y[0]) + '\n')
f.close()
f2.close()

def cost(w):
    pred = np.dot(X, w)
    return np.sqrt(((pred - Y) ** 2).mean(axis=None))

from autograd import grad
grad_cost= grad(cost)

W1 = 0
W2 = 0
b = 0
Weights = np.array([W1, W2, b]).reshape((3, 1))
i = 0
for i in range(0,100000):
    temp = Weights
    Weights = Weights - grad_cost(Weights)*0.01
print (cost(Weights), Weights)
print(i)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
pred = np.dot(X, Weights)
plt.plot(X_data,pred)
plt.plot(X_data, Y, '-')
plt.show()
```
