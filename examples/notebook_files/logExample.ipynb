```python
import matplotlib.pyplot as plt
```

```python
f = open('../txt_files/logEx.txt', 'r')
lines = [x.split() for x in f.readlines()]

X = [x[0] for x in lines]
Y = [x[1] for x in lines]
Ydiff = [x[2] for x in lines]
```

```python
plt.plot(X,Y,label='log')
plt.plot(X,Ydiff,label='diff log')
plt.savefig('../visuals/log.png')
plt.show()
```
